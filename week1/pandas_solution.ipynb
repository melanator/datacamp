```python
import pandas as pd
```

```python
df = pd.read_csv('yellow_tripdata_2021-01.csv')
```

```python
zones = pd.read_csv('taxi+_zone_lookup.csv')
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df
```

<b>Question 3: Count records</b><br>
How many taxi trips were there on January 15?

```python
df[df['tpep_pickup_datetime'].dt.floor('d')=="2021-01-15"].shape[0]
```

<b>Question 4: Largest tip for each day</b><br>
On which day it was the largest tip in January? (note: it's not a typo, it's "tip", not "trip")

```python
df[['tpep_pickup_datetime', 'tip_amount']].groupby(df['tpep_pickup_datetime'] \
                                                   .dt.floor('d')).max('tip_amount') \
                                                   .sort_values('tip_amount',ascending=False).head()
```

<b>Question 5: Most popular destination</b><br>
What was the most popular destination for passengers picked up in central park on January 14? Enter the zone name (not id). If the zone name is unknown (missing), write "Unknown"

```python
#Retrive id of Central Park
cp_zone = zones['LocationID'][zones['Zone']=='Central Park'].iloc[0]
```

```python
cp_zone
```

```python
#Dataframe of trips on January 14 from Central Park
cp_df = df[df['tpep_pickup_datetime'].dt.floor('d')=="2021-01-14"][df['PULocationID']==cp_zone]
```

```python
#Most frequent destination
freq_dest = cp_df.groupby(cp_df['DOLocationID']).count().sort_values('tpep_pickup_datetime',ascending=False).head().index[0]
print(freq_dest)
```

```python
zones[zones['LocationID']==freq_dest]
```

```python
#New column with pair PickUpID_DropOffID
df['PU_DO'] = df['PULocationID'].astype(str) + "_" + df['DOLocationID'].astype(str)
```
